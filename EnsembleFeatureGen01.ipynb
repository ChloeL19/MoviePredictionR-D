```python
# using this notebook to create feature data from AlexNet, inception, and other famous models --> going to create
# an ensemble method of sorts

from keras.applications.resnet50 import ResNet50
import cv2
from keras.models import Model
import numpy as np
import os
```

```python
model = ResNet50(include_top=False, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
print(model.summary())
```

```python
test_img = cv2.imread('../Desktop/frames/7/frame-00017.jpg')
print(test_img.shape)
```

```python
# loading data for one movie and storing all of it into a feature data array

resnet_feature_data = np.zeros([1, 212, 4096])
# for each file in the movie's frames folder
count = 0
for file in os.listdir('../Desktop/frames/7/'):
    filename = os.fsdecode(file)
    print(filename)
    if filename.endswith(".jpg"):
        # extract the image data
        test_img = cv2.imread(os.path.join('../Desktop/frames/7/', filename))
        data = (test_img/255)[np.newaxis, :, :, :]
        # send image data through the resnet model
        intermediate_output = model.predict(data)
        # reshape model output
        feature_data = intermediate_output.reshape([2*2048])
        # slot into the master away
        resnet_feature_data[0, count, :] = feature_data
        
        count = count + 1
        continue
    else:
        continue
    # extract the image data
    # feed it though the resnet model
    # slot output into the master array
print(resnet_feature_data.shape)
print(resnet_feature_data)
```

```python
# extracting feature data for all of the movies
# iterate though each movie in the folder
num_movies = 7

# set up master array
resnet_feature_data = np.zeros([7, 212, 4096])

for num in range(num_movies):
    count = 0
    for file in os.listdir('../Desktop/frames/{}/'.format(7+num)):
        filename = os.fsdecode(file)
        if filename.endswith(".jpg"):
            # extract the image data
            test_img = cv2.imread(os.path.join('../Desktop/frames/{}/'.format(7+num), filename))
            data = (test_img/255)[np.newaxis, :, :, :]
            # send image data through the resnet model
            intermediate_output = model.predict(data)
            # reshape model output
            feature_data = intermediate_output.reshape([2*2048])
            # slot into the master away
            resnet_feature_data[num, count, :] = feature_data
        
            count = count + 1
            continue
        else:
            continue
    print(num)
print(resnet_feature_data.shape)
print(resnet_feature_data)
```

```python
# save the feature data matrix
np.save('../Desktop/ResNet50_feature_data.npy', resnet_feature_data)
```
