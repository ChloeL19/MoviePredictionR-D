```python
# find some key variables, like the max number of seconds in each movie

VGG_data_dim = 4096
num_movies=7
# find the movie with the longest number of seconds
# set timesteps to zero
timesteps = 0
# iterate through all of the movies and compute num_secs/timesteps for each
# for each movie
for num in range(num_movies):
    if (7 + num) < 10:
        path = os.path.join("data_movies/MEDIAEVAL18_0{}.mp4".format(7 + num))
    else:
        path = os.path.join("data_movies/MEDIAEVAL18_{}.mp4".format(7 + num))
    vidcap = cv2.VideoCapture(path)
    # calculate the number of seconds
    frame_rate = math.floor(vidcap.get(cv2.CAP_PROP_FPS))
    # get total number of frames in the movie
    num_frames = vidcap.get(cv2.CAP_PROP_FRAME_COUNT)
    # get the number of seconds in the movie by dividing by frames per second 
    curr_timesteps = round(num_frames/frame_rate)
    # if the current timestep value is greater than timesteps
    if curr_timesteps > timesteps:
        # set timesteps to the curr_timestep value
        timesteps = curr_timesteps

```

```python
# will start by creating the labels matrix for the 7 movies
labels_matrix = np.zeros([num_movies, timesteps])

# for each movie number between and including 7 and 13
for num in range(num_movies):
    # create the appropriate path to the fear annotation data
    #print(num)
    if num+7 < 10:
        path = os.path.join('/Users/chloeloughridge/Desktop/AI Grant Project/Dataset/2018 Task Data/MEDIAEVAL18-DevSet-Part1-Fear-annotations/annotations/MEDIAEVAL18_0{}_Fear.txt'.format(7+num))
    if num+7 > 10:
        path = os.path.join('/Users/chloeloughridge/Desktop/AI Grant Project/Dataset/2018 Task Data/MEDIAEVAL18-DevSet-Part1-Fear-annotations/annotations/MEDIAEVAL18_{}_Fear.txt'.format(7+num))
    # create a one-hot vector
    y_data = fear_oneHot(timesteps, path)
    # add this one-hot vector to y_train_data
    labels_matrix[num, :] = y_data
    
print(labels_matrix[4])
```
